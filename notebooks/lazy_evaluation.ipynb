# Lazy evalutation in Dask

Dask delays execution until it is specifically asked to do so. It creates instead a task graph to optimize computation

```python
# First, create a client
from dask.distributed import Client
```

```python
client = Client(n_workers=4, threads_per_worker=1)
client
```

```python
# Contruct a big random array
import dask.array as da
```

```python
SIZE = 10000
CHUNK_SIZE = 1000
arr = da.random.random((SIZE,SIZE), (CHUNK_SIZE,CHUNK_SIZE))
```

```python
# The result is not computed yet
result = arr.mean()
```

```python
# You can visualize the resulting task graph (make sure to have Graphviz installed)
result.visualize()
```
